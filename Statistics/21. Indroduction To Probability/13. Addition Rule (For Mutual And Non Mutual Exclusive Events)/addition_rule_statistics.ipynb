## üîπ 1. What is the Addition Rule?

The **Addition Rule** in probability is used to find the probability that **at least one** of two events will occur.

### General Formula:
For any two events A and B,
\[
P(A \cup B) = P(A) + P(B) - P(A \cap B)
\]

---

## üîπ 2. Mutually Exclusive Events

### ‚úÖ Definition:
- Two events are **mutually exclusive** if they **cannot happen at the same time**.
- That means:  
  \[
  P(A \cap B) = 0
  \]

### üî∏ Formula for Mutually Exclusive Events:
\[
P(A \cup B) = P(A) + P(B)
\]

### üìå Example:
Let‚Äôs say:
- P(A) = probability of rolling a 1 on a die = 1/6
- P(B) = probability of rolling a 2 on a die = 1/6

These two are **mutually exclusive**, since a single die can‚Äôt show both 1 and 2 at once.

### ‚úÖ Output:
\[
P(A \cup B) = 1/6 + 1/6 = 1/3
\]

---

## üîπ 3. Non-Mutually Exclusive Events

### ‚úÖ Definition:
- Two events are **non-mutually exclusive** if they **can both occur** at the same time.
- That means:  
  \[
  P(A \cap B) \ne 0
  \]

### üî∏ Formula:
\[
P(A \cup B) = P(A) + P(B) - P(A \cap B)
\]

### üìå Example:
- Event A: A student takes Math (P(A) = 0.5)
- Event B: A student takes Science (P(B) = 0.6)
- Event A ‚à© B: Student takes both (P(A ‚à© B) = 0.3)

### ‚úÖ Output:
\[
P(A \cup B) = 0.5 + 0.6 - 0.3 = 0.8
\]

---

## üîÅ Comparison Table

| Feature                        | Mutually Exclusive           | Non-Mutually Exclusive         |
|-------------------------------|------------------------------|--------------------------------|
| Can occur together?           | ‚ùå No                        | ‚úÖ Yes                         |
| Intersection \( P(A ‚à© B) \) | 0                            | Not necessarily 0              |
| Formula                       | \( P(A) + P(B) \)          | \( P(A) + P(B) - P(A ‚à© B) \) |

---

## üîß Use in Python

```python
# Mutually exclusive
P_A = 1/6
P_B = 1/6
P_A_or_B = P_A + P_B
print("Mutually Exclusive:", round(P_A_or_B, 2))  # Output: 0.33

# Non-mutually exclusive
P_A = 0.5
P_B = 0.6
P_A_and_B = 0.3
P_A_or_B = P_A + P_B - P_A_and_B
print("Non-Mutually Exclusive:", round(P_A_or_B, 2))  # Output: 0.8
```

---

## üìù Summary

- Use the **addition rule** to find the probability that **at least one** event happens.
- If events are **mutually exclusive**, use:  
  \( P(A \cup B) = P(A) + P(B) \)
- If events are **not mutually exclusive**, subtract their intersection:  
  \( P(A \cup B) = P(A) + P(B) - P(A \cap B) \)
- Always check if events can happen together or not.

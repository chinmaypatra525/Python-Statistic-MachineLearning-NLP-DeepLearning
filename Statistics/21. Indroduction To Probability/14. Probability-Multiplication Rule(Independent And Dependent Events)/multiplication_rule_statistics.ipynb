# ✖️ Statistics Notes: Multiplication Rule (Independent and Dependent Events)

---

## 🔹 1. What is the Multiplication Rule?

The **Multiplication Rule** is used to calculate the probability of **two events happening together**, i.e., the **intersection** of events \( A \) and \( B \):

\[
P(A \cap B) = P(A) \cdot P(B|A)
\]

Where:
- \( P(A) \) is the probability of event A
- \( P(B|A) \) is the probability of event B **given** that A has occurred

---

## 🔹 2. Independent Events

### ✅ Definition:
Two events are **independent** if the occurrence of one **does not affect** the probability of the other.

### 🔸 Rule:
\[
P(A \cap B) = P(A) \cdot P(B)
\]

### 📌 Example:
- Event A: Tossing a coin and getting heads → \( P(A) = 0.5 \)
- Event B: Rolling a die and getting a 3 → \( P(B) = \frac{1}{6} \)

Since these are unrelated actions, they are independent.

### ✅ Calculation:
\[
P(A \cap B) = 0.5 \times \frac{1}{6} = \frac{1}{12} \approx 0.083
\]

---

## 🔹 3. Dependent Events

### ✅ Definition:
Two events are **dependent** if the occurrence of one **affects the probability** of the other.

### 🔸 Rule:
\[
P(A \cap B) = P(A) \cdot P(B|A)
\]

### 📌 Example:
- A: Drawing an Ace from a 52-card deck → \( P(A) = \frac{4}{52} \)
- B: Drawing a King **after** the Ace is drawn, **without replacement** → \( P(B|A) = \frac{4}{51} \)

These events are dependent because the second draw changes based on the first.

### ✅ Calculation:
\[
P(A \cap B) = \frac{4}{52} \cdot \frac{4}{51} = \frac{16}{2652} \approx 0.006
\]

---

## 🔁 Comparison Table

| Feature                        | Independent Events             | Dependent Events                   |
|-------------------------------|--------------------------------|------------------------------------|
| Affect each other?            | ❌ No                          | ✅ Yes                             |
| Formula                       | \( P(A) \cdot P(B) \)          | \( P(A) \cdot P(B|A) \)            |
| Example                       | Coin toss & die roll           | Drawing cards without replacement  |

---

## 🔧 Python Example

```python
# Independent Events
P_A = 0.5      # Coin flip heads
P_B = 1/6      # Die roll 3
P_both_indep = P_A * P_B
print("Independent Events:", round(P_both_indep, 3))  # Output: 0.083

# Dependent Events
P_A = 4/52     # Drawing an Ace
P_B_given_A = 4/51  # Drawing a King after drawing an Ace (without replacement)
P_both_dep = P_A * P_B_given_A
print("Dependent Events:", round(P_both_dep, 4))  # Output: ~0.0062


## 📝 Summary

The **Multiplication Rule** is used when calculating the probability of **A and B happening together**.

- If events are **independent**, use:
  \[
  P(A \cap B) = P(A) \cdot P(B)
  \]

- If events are **dependent**, use:
  \[
  P(A \cap B) = P(A) \cdot P(B|A)
  \]

Always check whether one event influences the other (e.g., drawing with or without replacement).

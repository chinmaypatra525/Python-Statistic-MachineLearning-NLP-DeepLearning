# Bayes’ Theorem

---

## What is Bayes’ Theorem?

Bayes’ Theorem is a fundamental rule in probability theory that describes how to update the probability of a hypothesis based on new evidence.

It relates the conditional and marginal probabilities of random events.



## Formula

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$


Where:

| Symbol     | Meaning                                                       |
|------------|---------------------------------------------------------------|
| \(P(A\|B)\) | Posterior probability: Probability of event A given B occurred (updated probability). |
| \(P(B\|A)\) | Likelihood: Probability of event B given that A is true.      |
| \(P(A)\)   | Prior probability: Initial probability of event A before observing B. |
| \(P(B)\)   | Marginal likelihood (evidence): Total probability of event B.  |




## Explanation

- **Prior \(P(A)\):** Your initial belief about event A before seeing evidence B.
- **Likelihood \(P(B|A)\):** How likely is the evidence B if A is true.
- **Marginal Likelihood \(P(B)\):** Overall probability of observing evidence B under all hypotheses.
- **Posterior \(P(A|B)\):** Updated probability of A after observing B.



## Real-Life Example

**Scenario:**  
A medical test for a disease is 99% accurate (both true positive and true negative). The disease prevalence is 0.5%. Given a positive test result, what is the probability that the person actually has the disease?

- Let:  
  - \(A\) = person has the disease  
  - \(B\) = test is positive

- Known values:  
  - $P(A) = 0.005$ (disease prevalence)  
  - $P(B|A) = 0.99$ (true positive rate)  
  - $P(B|\neg A) = 0.01$ (false positive rate)


- Calculate \(P(B)\):  
$$
P(B) = P(B|A)P(A) + P(B|\neg A)P(\neg A) = (0.99)(0.005) + (0.01)(0.995) = 0.0149
$$


- Apply Bayes’ theorem:  
$$
P(A|B) = \frac{(0.99)(0.005)}{0.0149} \approx 0.33
$$


**Interpretation:**  
Even after testing positive, there’s only about a 33% chance the person actually has the disease, highlighting the impact of low prevalence.



## Summary

| Term             | Description                              |
|------------------|------------------------------------------|
| Prior Probability | Initial probability before evidence      |
| Likelihood       | Probability of evidence given hypothesis  |
| Marginal Likelihood| Overall probability of evidence          |
| Posterior Probability | Updated probability after evidence     |

Bayes’ Theorem provides a way to update beliefs in light of new data and is widely used in statistics, machine learning, and decision making.

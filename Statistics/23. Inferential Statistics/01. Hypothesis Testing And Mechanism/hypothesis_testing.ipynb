## üß† What is Hypothesis Testing?

**Hypothesis Testing** is a statistical method used to make decisions about a population parameter based on a sample.

It helps determine whether the evidence from a sample supports a specific claim or not.



## üéØ Goal

To test whether a **claim (hypothesis)** about a population parameter is likely to be true based on sample data.



## üß™ Key Terms

| Term | Description |
|------|-------------|
| **Null Hypothesis (H‚ÇÄ)** | The default assumption ‚Äî no effect or no difference. |
| **Alternative Hypothesis (H‚ÇÅ or H‚Çê)** | The claim we want to test for ‚Äî there **is** an effect or difference. |
| **Test Statistic** | A number calculated from sample data used to decide whether to reject H‚ÇÄ. |
| **p-value** | The probability of getting the observed results, assuming H‚ÇÄ is true. |
| **Significance Level (Œ±)** | The threshold for rejecting H‚ÇÄ. Common values: 0.05, 0.01. |
| **Type I Error (Œ±)** | Rejecting H‚ÇÄ when it is actually true (false positive). |
| **Type II Error (Œ≤)** | Failing to reject H‚ÇÄ when H‚ÇÅ is actually true (false negative). |



## ‚öôÔ∏è Mechanism of Hypothesis Testing

1. **State the Hypotheses**  
   - Null Hypothesis (H‚ÇÄ): e.g., Œº = 50  
   - Alternative Hypothesis (H‚ÇÅ): e.g., Œº ‚â† 50

2. **Choose the Significance Level (Œ±)**  
   - Commonly 0.05 (5% chance of Type I error)

3. **Collect Sample Data**  
   - Example: Take a random sample and compute the test statistic.

4. **Compute Test Statistic and p-value**  
   - Depends on the test (Z-test, t-test, etc.)

5. **Make a Decision**  
   - If **p-value ‚â§ Œ±** ‚Üí Reject H‚ÇÄ  
   - If **p-value > Œ±** ‚Üí Fail to reject H‚ÇÄ

6. **State Conclusion in Context**



## ‚úÖ Example: One-Sample t-Test

### Scenario:
A battery manufacturer claims their battery lasts **on average 10 hours**. You want to test this claim.

- Sample Mean = 9.6 hours  
- Population Standard Deviation unknown  
- Sample Size = 25  
- Œ± = 0.05

### Hypotheses:

- H‚ÇÄ: Œº = 10  
- H‚ÇÅ: Œº ‚â† 10

### Result:
- t-statistic = -2.5  
- p-value = 0.02

### Decision:
- p-value (0.02) < Œ± (0.05) ‚áí **Reject H‚ÇÄ**

### Conclusion:
There is significant evidence that the battery life differs from 10 hours.



## üìâ Common Statistical Tests

| Test | Use Case |
|------|----------|
| **Z-test** | Population std. dev. known, large sample |
| **t-test** | Population std. dev. unknown, small sample |
| **Chi-square test** | Categorical variables |
| **ANOVA** | Comparing more than two group means |
| **Proportion test** | Testing population proportions |



## üí° Key Insights

- Hypothesis testing is essential for **data-driven decision making**.
- It does **not prove** a hypothesis ‚Äî it measures **evidence** against the null.
- Lower p-values mean **stronger evidence** against the null.
- Always consider **effect size and confidence intervals**, not just p-values.

(Shows critical region, significance level, and power of the test)



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0           # Mean under H0
sigma = 1        # Standard deviation
alpha = 0.05     # Significance level
z_crit = norm.ppf(1 - alpha/2)  # Two-tailed critical z-value

# Z range
z = np.linspace(-4, 4, 1000)
pdf = norm.pdf(z, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(z, pdf, label='Normal Distribution (H‚ÇÄ)', color='black')

# Fill critical regions
plt.fill_between(z, 0, pdf, where=(z <= -z_crit), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(z, 0, pdf, where=(z >= z_crit), color='red', alpha=0.5, label='Rejection Region (Right)')

# Fill non-rejection region
plt.fill_between(z, 0, pdf, where=(z > -z_crit) & (z < z_crit), color='green', alpha=0.3, label='Fail to Reject H‚ÇÄ')

# Annotate
plt.axvline(x=-z_crit, color='red', linestyle='--', linewidth=1)
plt.axvline(x=z_crit, color='red', linestyle='--', linewidth=1)
plt.text(-z_crit, 0.02, f'-z‚Çê/‚ÇÇ = {-z_crit:.2f}', ha='center', color='red')
plt.text(z_crit, 0.02, f'z‚Çê/‚ÇÇ = {z_crit:.2f}', ha='center', color='red')

# Optional test statistic line (e.g., z = 2)
test_stat = 2.0
plt.axvline(x=test_stat, color='blue', linestyle=':', linewidth=2, label=f'Test Statistic (z = {test_stat})')

# Labels and title
plt.title('Two-Tailed Hypothesis Testing Curve')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# üéØ Type I and Type II Errors in Hypothesis Testing

Understanding the **errors** in hypothesis testing is crucial for drawing valid conclusions from data.

---

## üß† Hypothesis Recap

- **Null Hypothesis (H‚ÇÄ)**: No effect, no difference, or the status quo.
- **Alternative Hypothesis (H‚ÇÅ)**: There is an effect or a difference.



## ‚ùå Type I Error (False Positive)

### üîç Definition:
Rejecting the **null hypothesis** when it is actually **true**.

### üß™ Example:
You're testing a drug that **does not work**, but your test concludes that it **does**.

### üìå Probability:
The probability of making a Type I error is denoted by **Œ± (alpha)**.

### üõ°Ô∏è Control:
Set a significance level (commonly 0.05) to limit the chance of this error.



## ‚ùå Type II Error (False Negative)

### üîç Definition:
Failing to reject the **null hypothesis** when it is actually **false**.

### üß™ Example:
The drug **does work**, but your test concludes that it **doesn't**.

### üìå Probability:
The probability of making a Type II error is denoted by **Œ≤ (beta)**.

### üõ°Ô∏è Control:
Increase **sample size**, reduce variance, or raise the **significance level**.



## üîÑ Comparison Table

| Feature              | Type I Error                    | Type II Error                    |
|----------------------|----------------------------------|----------------------------------|
| Symbol               | Œ± (alpha)                       | Œ≤ (beta)                         |
| What happens         | Reject H‚ÇÄ when it's true        | Fail to reject H‚ÇÄ when it's false |
| Also known as        | False Positive                  | False Negative                   |
| Controlled by        | Significance level              | Power of the test (1 - Œ≤)        |
| Real-world example   | Innocent person convicted       | Guilty person acquitted          |



## ‚öñÔ∏è Balancing the Errors

- **Lowering Œ±** reduces the chance of **Type I Error**, but **increases** the chance of **Type II Error**.
- **Power of a test** = 1 ‚àí Œ≤: the probability that the test **correctly rejects** a false null hypothesis.







## ‚úÖ Key Insights

- Every hypothesis test carries a risk of **error**.
- Choosing Œ± = 0.05 means you're **okay with a 5% chance** of a Type I error.
- Type I and Type II errors are **inversely related**‚Äîreducing one typically increases the other.

## üìâ  Example

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Set seaborn style
sns.set(style="whitegrid")

# Parameters
mu_null = 0         # Mean under null hypothesis H0
mu_alt = 1          # Mean under alternative hypothesis H1
sigma = 1           # Standard deviation
alpha = 0.05        # Significance level

# Generate x values
x = np.linspace(-4, 5, 1000)

# Distributions
null_dist = stats.norm(mu_null, sigma)
alt_dist = stats.norm(mu_alt, sigma)

# Critical value for right-tailed test
critical_value = stats.norm.ppf(1 - alpha)

plt.figure(figsize=(10, 6))

# Plot distributions with seaborn color palette
sns.lineplot(x=x, y=null_dist.pdf(x), label='H‚ÇÄ: Null Hypothesis (Œº=0)', color=sns.color_palette("tab10")[0])
sns.lineplot(x=x, y=alt_dist.pdf(x), label='H‚ÇÅ: Alternative Hypothesis (Œº=1)', color=sns.color_palette("tab10")[1])

# Shade Type I error (Œ±)
x_alpha = np.linspace(critical_value, 5, 200)
plt.fill_between(x_alpha, null_dist.pdf(x_alpha), color=sns.color_palette("tab10")[0], alpha=0.3, label='Type I Error (Œ±)')

# Shade Type II error (Œ≤)
x_beta = np.linspace(-4, critical_value, 200)
plt.fill_between(x_beta, alt_dist.pdf(x_beta), color=sns.color_palette("tab10")[1], alpha=0.3, label='Type II Error (Œ≤)')

# Critical value line
plt.axvline(x=critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

# Labels and title
plt.title('Visualization of Type I and Type II Errors with Seaborn')
plt.xlabel('Test Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Step-by-Step Breakdown

1. **Import Libraries**  
   - `numpy` for numerical calculations and generating ranges.  
   - `matplotlib.pyplot` for plotting.  
   - `seaborn` for nicer plot aesthetics.  
   - `scipy.stats` for working with normal distributions.

2. **Set Plot Style**  
   Use `seaborn.set(style="whitegrid")` to apply a clean background and grid.

3. **Define Parameters**  
   - `mu_null = 0`: Mean under the null hypothesis \(H_0\).  
   - `mu_alt = 1`: Mean under the alternative hypothesis \(H_1\).  
   - `sigma = 1`: Standard deviation (same for both distributions).  
   - `alpha = 0.05`: Significance level, the threshold for Type I error.

4. **Generate Values for X-axis**  
   `x` contains 1000 points evenly spaced from -4 to 5, representing possible values of the test statistic.

5. **Create Normal Distributions**  
   - `null_dist` is the normal distribution under \(H_0\) (mean = 0).  
   - `alt_dist` is the normal distribution under \(H_1\) (mean = 1).

6. **Calculate Critical Value**  
   The critical cutoff for rejecting \(H_0\) in a right-tailed test is calculated with  
   `stats.norm.ppf(1 - alpha)`.

7. **Initialize Plot**  
   Create a figure with size 10x6 inches.

8. **Plot Probability Density Functions (PDFs)**  
   - Plot \(H_0\) distribution in blue.  
   - Plot \(H_1\) distribution in red.

9. **Shade Type I Error Region (Œ±)**  
   Shade the area to the right of the critical value under the null distribution.  
   This represents rejecting \(H_0\) when it is true (false positive).

10. **Shade Type II Error Region (Œ≤)**  
    Shade the area to the left of the critical value under the alternative distribution.  
    This represents failing to reject \(H_0\) when \(H_1\) is true (false negative).

11. **Draw Critical Value Line**  
    Add a dashed vertical line at the critical value to indicate the rejection threshold.

12. **Add Labels and Show Plot**  
    Add title, axis labels, legend, and grid.  
    Display the plot.



### Interpretation

- The **blue shaded region** represents the **Type I error (Œ±)**: rejecting the null hypothesis incorrectly.  
- The **red shaded region** represents the **Type II error (Œ≤)**: failing to reject the null hypothesis when the alternative is true.  
- The **vertical dashed line** marks the decision boundary based on the significance level.  
- This visualization helps understand the trade-off between Type I and Type II errors in hypothesis testing.

## üß† What is Poisson Distribution?

The **Poisson Distribution** models the probability of a given number of events occurring in a **fixed interval of time or space**, given that these events occur **independently** and at a **constant average rate**.

> üìå It is useful when you're counting occurrences of **rare events**.



## ‚úÖ Key Characteristics

- Events are **independent** of each other
- Events happen at a **constant average rate** \( \lambda \)
- Two events can't happen at **exactly the same time**
- The probability of an event occurring in a small interval is proportional to the **length of the interval**

## üí° Key Insights

‚úÖ The **Poisson distribution** is often used to model **rare events**.

‚è±Ô∏è It assumes events occur **independently** and at a **constant average rate**.

üî¢ **Mean and Variance are equal**:  



$\mu = \sigma^2 = \lambda$



üìâ The distribution is **right-skewed for small** \( \lambda \); becomes more **symmetric as \( \lambda \) increases**.

üîÅ **Poisson is a limiting case** of the Binomial Distribution as:  


$$
n \to \infty,\quad p \to 0,\quad \text{and} \quad np = \lambda
$$





## üî¢ Probability Mass Function (PMF)

The PMF of a Poisson-distributed random variable \( X \) is:

$$
P(X = x) = \frac{\lambda^x e^{-\lambda}}{x!}, \quad x = 0, 1, 2, \dots
$$


Where:
- $\lambda$ = average number of events  
- $e$ = Euler‚Äôs number (~2.718)  
- $x!$ = factorial of $x$







## üõ† Python Code:


```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

lambda_ = 3
x = range(10)
probs = poisson.pmf(x, lambda_)

plt.bar(x, probs, color='green')
plt.title('Poisson Distribution (Œª=3)')
plt.xlabel('k (events)')
plt.ylabel('P(X = k)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lam = 3  # Average number of events

# X values
x = np.arange(0, 11)
pmf = poisson.pmf(x, mu=lam)

# Plot
plt.bar(x, pmf, color='skyblue')
plt.title(f"Poisson Distribution (Œª={lam})")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True, axis='y')
plt.show()

# Print PMF values
for val, prob in zip(x, pmf):
    print(f"P(X = {val}) = {prob:.4f}")
```

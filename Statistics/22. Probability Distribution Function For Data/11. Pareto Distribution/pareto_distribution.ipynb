## 🧠 What is Pareto Distribution?

The **Pareto Distribution** is a **power-law probability distribution** used to describe many types of **observable phenomena in economics, finance, and social sciences**. It is named after **Vilfredo Pareto**, who observed that 80% of wealth was owned by 20% of the population — known as the **80/20 rule**.

It models the probability that a variable (like wealth, file size, city population) is **greater than a certain value**.



## 🧾 Real-Life Examples

| Scenario                                | Description                          |
|-----------------------------------------|--------------------------------------|
| Wealth distribution                     | Small % of people own most wealth    |
| File sizes on the internet              | Few large files, many small ones     |
| City populations                        | Few megacities, many small towns     |
| Insurance claim sizes                   | Few large claims, many small ones    |
| Word frequencies in language            | Few words used frequently            |



## ✅ When to Use

- 📏 When modeling systems with **scale invariance**
- 🐳 When data exhibits **long tail** or **rich-get-richer behavior**
- 💼 Applicable in **economics, ecology, geophysics, insurance**, etc.



## 📊 Properties

| Property               | Formula / Description                                               |
|------------------------|---------------------------------------------------------------------|
| Type                   | Continuous                                                          |
| Support                | \( x \geq x_m \) (minimum value)                                   |
| Shape Parameter        | \( \alpha > 0 \)                                                    |
| PDF                    | \( f(x) = \frac{\alpha x_m^\alpha}{x^{\alpha + 1}} \)              |
| CDF                    | \( F(x) = 1 - \left( \frac{x_m}{x} \right)^\alpha \)               |
| Mean                   | \( \mu = \frac{\alpha x_m}{\alpha - 1} \), if \( \alpha > 1 \)      |
| Variance               | \( \sigma^2 = \frac{\alpha x_m^2}{(\alpha - 1)^2 (\alpha - 2)} \), if \( \alpha > 2 \) |
| Skewness               | Heavy right-skewed                                                 |
| Memoryless Property    | No                                                                 |



## 🧠 Summary

| **Support**      | \( x \geq x_m \) |
| **Skewness**     | Heavy right-tail |
| **Applications** | Wealth, size distribution, file sizes |



## 💡 Key Insights

- 💰 Models **"rich-get-richer"** and **heavy-tail** phenomena.
- 🌐 Often used in:
  - **Economics** (e.g., income distribution)
  - **Network modeling**
  - **Risk assessment**
- ⚠️ Unlike Normal Distribution, **Pareto has infinite variance if α ≤ 2**.
- 🔗 Closely related to the **Power Law Distribution**, but with a **specific lower bound**.



## 🔍 Notes

- 📉 When plotted on a **log-log scale**, Pareto forms a **straight line** (indicative of power-law behavior).
- ❗ Useful in scenarios where **mean and variance may not exist** (for low values of α).



## 🔢 Probability Density Function (PDF)

\[
f(x; \alpha, x_m) =
\begin{cases}
\frac{\alpha x_m^\alpha}{x^{\alpha + 1}}, & x \geq x_m \\
0, & x < x_m
\end{cases}
\]

Where:
- \( \alpha \) = shape parameter
- \( x_m \) = scale parameter (minimum value)



## 📌 Interpretation

- 🔽 **Larger α → faster drop-off** (less extreme events)  
- 📈 **Smaller α → heavier tail** (more extreme values)  
- ♾️ **Pareto explains self-similarity**, where distributions look the same on different scales.



## 🛠 Python Code:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

b = 2.62
x = np.linspace(1, 5, 1000)
y = pareto.pdf(x, b)

plt.plot(x, y, color='magenta')
plt.title('Pareto Distribution (α=2.62)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

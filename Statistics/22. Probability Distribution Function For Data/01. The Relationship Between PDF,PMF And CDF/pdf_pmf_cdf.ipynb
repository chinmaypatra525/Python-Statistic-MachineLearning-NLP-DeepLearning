# 📊 Probability Distribution Function (PDF) – Statistics Notes

---

## 🔹 What is a Probability Distribution Function?

A **Probability Distribution Function (PDF)** describes how the probabilities are distributed over the values of a **continuous random variable**.

Unlike PMF (used for discrete variables), a PDF gives the probability **density**, not the exact probability.

---

### 🧠 Key Concepts

- For a **continuous random variable**, the **probability that it takes on any single value is zero**, i.e., $P(X = x) = 0$
- Instead, we calculate:

  $$
  P(a \leq X \leq b) = \int_a^b f(x)\, dx
  $$

  where $f(x)$ is the PDF.

- A PDF must satisfy:
  1. $f(x) \geq 0$ for all $x$
  2. The total area under the curve is 1:

     $$
     \int_{-\infty}^{\infty} f(x)\, dx = 1
     $$



### 🔸 Example: Uniform Distribution

Let $X \sim U(0, 1)$ be a uniform distribution.

- The PDF is:

  $$
  f(x) =
  \begin{cases}
  1, & \text{if } 0 \leq x \leq 1 \\
  0, & \text{otherwise}
  \end{cases}
  $$

- Probability that $X$ lies between 0.2 and 0.5:

  $$
  P(0.2 \leq X \leq 0.5) = \int_{0.2}^{0.5} 1 \, dx = 0.5 - 0.2 = 0.3
  $$


### 🔸 Example: Normal Distribution

Let $X \sim N(\mu, \sigma^2)$, a normal distribution with:
- Mean $\mu = 0$
- Standard deviation $\sigma = 1$

The PDF is:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2 / 2}
$$


Because this PDF doesn't have a simple closed-form integral, we use a **Z-table** or **computational methods** to evaluate probabilities.



### 🔧 Python Example


````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define range and PDF
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)

# Plot PDF
plt.plot(x, pdf, label='Standard Normal PDF')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()

# Probability between -1 and 1
p = norm.cdf(1) - norm.cdf(-1)
print("P(-1 ≤ X ≤ 1) =", round(p, 4))  # Output: ~0.6826
````

# 📊 Types of Probability Distribution Functions (PDFs)

Probability Distribution Functions describe how probabilities are distributed over values of a random variable. These are categorized into:

- **Discrete Probability Distributions**
- **Continuous Probability Distributions**

---

## 🔹 1. Discrete Probability Distributions (PMF - Probability Mass Function)

A discrete distribution is one where the random variable can take **finite or countably infinite** values.

### ✅ Common Types:



#### 1.1 Bernoulli Distribution

- Describes a single experiment with two outcomes: **success (1)** and **failure (0)**
- Parameter: $p$, the probability of success
- PMF:

  $$
  P(X = x) = p^x (1-p)^{1-x}, \quad x \in \{0, 1\}
  $$



#### 1.2 Binomial Distribution

- Number of successes in **n** independent Bernoulli trials
- Parameters: $n$, $p$
- PMF:

  $$
  P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
  $$



#### 1.3 Poisson Distribution

- Counts number of events in a fixed interval (time, space)
- Used when events occur **independently**
- Parameter: $\lambda$ (mean rate)
- PMF:

  $$
  P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
  $$



## 🔹 2. Continuous Probability Distributions (PDF - Probability Density Function)

A continuous distribution describes a variable that can take **any value** in a range.

### ✅ Common Types:



#### 2.1 Uniform Distribution

- All intervals of same length have equal probability
- Parameters: $a$, $b$
- PDF:

  $$
  f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
  $$



#### 2.2 Normal (Gaussian) Distribution

- Bell-shaped curve, widely used
- Parameters: $\mu$ (mean), $\sigma$ (std deviation)
- PDF:

  $$
  f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
  $$



#### 2.3 Exponential Distribution

- Describes time between events in a Poisson process
- Parameter: $\lambda$
- PDF:

  $$
  f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
  $$


#### 2.4 Beta Distribution

- Continuous distribution defined on $[0, 1]$
- Parameters: $\alpha$, $\beta$
- Flexible shape based on parameters
- Used in Bayesian statistics



#### 2.5 Gamma Distribution

- Generalized form of exponential distribution
- Parameters: shape $k$, rate $\lambda$
- PDF:

  $$
  f(x) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{\Gamma(k)}, \quad x \geq 0
  $$



#### 🔁 Summary Table

| Distribution     | Type       | Parameters         | Application Example               |
|------------------|------------|--------------------|-----------------------------------|
| Bernoulli         | Discrete   | $p$                | Single coin toss                  |
| Binomial          | Discrete   | $n, p$             | Number of heads in $n$ coin tosses |
| Poisson           | Discrete   | $\lambda$          | Calls per minute at a call center |
| Uniform           | Continuous | $a, b$             | Random number generator           |
| Normal            | Continuous | $\mu, \sigma$      | Heights of people                 |
| Exponential       | Continuous | $\lambda$          | Time until next earthquake        |
| Beta              | Continuous | $\alpha, \beta$    | Probabilities in Bayesian models  |
| Gamma             | Continuous | $k, \lambda$       | Insurance claim sizes             |



## 📝 Summary

- Discrete distributions have **PMFs**, Continuous distributions have **PDFs**
- Normal and Binomial are the most common for real-world modeling
- Understand the context (e.g., "count of events" vs. "time between events") to choose the right distribution


# 📊 Statistics Notes: PDF, PMF, and CDF with Examples and Summary

---

## 🔹 1. PMF (Probability Mass Function)

### ✅ Definition:

* PMF is used for **discrete random variables**.
* It gives the probability that a discrete random variable is exactly equal to some value.

#### 🧠 Properties:

* $P(X = x) \geq 0$
* $\sum P(X = x_i) = 1$ for all $x_i$ in the domain of $X$

#### 📌 Example:

Consider a fair 6-sided die.

* The PMF is:

  * $P(X=1) = \frac{1}{6}$
  * $P(X=2) = \frac{1}{6}$
  * ...
  * $P(X=6) = \frac{1}{6}$

✅ Output:
Each outcome from 1 to 6 has equal probability of 0.1667.



## 🔹 2. PDF (Probability Density Function)

### ✅ Definition:

* PDF is used for **continuous random variables**.
* It describes the **relative likelihood** of a random variable taking on a particular value.

#### 🧠 Properties:

* $f(x) \geq 0$
* $\int_{-\infty}^{\infty} f(x) dx = 1$
* The **probability** is given by the **area under the curve**:

  * $P(a \leq X \leq b) = \int_a^b f(x) dx$

#### 📌 Example:

A standard normal distribution has:

* Mean = 0, Std Dev = 1
* PDF: $f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$

✅ Output:
Probability that $X$ is between -1 and 1:

* $\approx 68.27\%$



## 🔹 3. CDF (Cumulative Distribution Function)

#### ✅ Definition:

* The CDF gives the **probability that the variable takes a value less than or equal to a certain value**.
* Applicable to both **discrete and continuous** distributions.

#### 🧠 Properties:

* $F(x) = P(X \leq x)$
* $\lim_{x \to -\infty} F(x) = 0$
* $\lim_{x \to \infty} F(x) = 1$
* Non-decreasing and right-continuous.

#### 📌 Example:

For a discrete variable like die roll:

* $F(3) = P(X \leq 3) = P(1) + P(2) + P(3) = \frac{3}{6} = 0.5$

For a normal distribution:

* $F(0) = 0.5$
* $F(1) \approx 0.8413$

✅ Output:
CDF at x=1 in standard normal distribution is 84.13%.



## 🔁 Comparison Table

| Feature       | PMF                    | PDF                                     | CDF                                  |
| ------------- | ---------------------- | --------------------------------------- | ------------------------------------ |
| Variable Type | Discrete               | Continuous                              | Both                                 |
| Outputs       | Probability at a point | Density at a point (area = probability) | Cumulative probability up to a point |
| Total         | Sum = 1                | Integral = 1                            | Ranges from 0 to 1                   |
| Graph         | Bar chart              | Smooth curve                            | Step or smooth increasing curve      |



## ✅ Summary

* **PMF**: Exact probabilities for discrete variables (e.g., dice, coin toss).
* **PDF**: Probability density (not exact value) for continuous variables (e.g., height, weight).
* **CDF**: Cumulative probability — total probability up to a certain value.
* Use area under curve (PDF) or summing (PMF) to find actual probabilities.



## 🔧 Use in Python:

```python
from scipy.stats import norm, randint

# PMF for discrete variable
print(randint.pmf(3, 1, 7))  # Output: 0.1667

# PDF for standard normal
print(norm.pdf(0))  # Output: 0.3989

# CDF for standard normal
print(norm.cdf(1))  # Output: ~0.8413
```


# 📈 Area Under the Curve (AUC) – Statistics & Machine Learning

---

### 🔹 What is Area Under the Curve?

In statistics and probability, the **Area Under the Curve (AUC)** refers to the **area under a probability density function (PDF)** or a **curve representing model performance**, such as in ROC analysis.

It can mean:

1. **Probability between values** for **continuous distributions**  
2. **Model evaluation metric** (like ROC AUC) in **classification tasks**



### 📊 1. AUC in Probability Distributions

For **continuous random variables**, the area under the **PDF** between two values $a$ and $b$ gives the **probability** that the variable falls within that range.

$$
P(a \leq X \leq b) = \int_a^b f(x)\, dx
$$

#### ✅ Properties:
- Total area under the entire curve:  
  $$
  \int_{-\infty}^{\infty} f(x)\, dx = 1
  $$
- Area between two points gives the probability of that range
- PDF values can be > 1, but **area** always ≤ 1

#### 📌 Example (Normal Distribution):

```python
from scipy.stats import norm

# Area under standard normal curve between -1 and 1
p = norm.cdf(1) - norm.cdf(-1)
print("P(-1 ≤ X ≤ 1) =", round(p, 4))


## 1. Plot Area Under Normal PDF Curve

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Define bounds for shaded area
a, b = -1, 1
x_fill = np.linspace(a, b, 1000)
y_fill = norm.pdf(x_fill)

# Plot the normal curve
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal PDF', color='blue')
plt.fill_between(x_fill, y_fill, alpha=0.3, color='skyblue', label=f'Area between {a} and {b}')
plt.title('Area Under the Curve: Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
````

## 2. Plot ROC Curve with AUC

````python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Sample data
y_true = [0, 0, 1, 1]
y_scores = [0.1, 0.4, 0.35, 0.8]

# ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
auc = roc_auc_score(y_true, y_scores)

# Plot
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, marker='o', label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve with AUC')
plt.legend()
plt.grid(True)
plt.show()
````

# 📈 Cumulative Probability and AUC (CDF)

---

### 🔹 What is Cumulative Probability?

The cumulative probability up to a point $x$ is given by:

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
$$

Where:
- $f(t)$ is the PDF  
- $F(x)$ is the CDF



### 🔹 Visual Meaning of AUC in CDF

- The area under the PDF curve from $-\infty$ to $x$ is the **cumulative probability**
- CDF is a **monotonic increasing** function that ranges from 0 to 1



### 🔹 CDF Properties

- $F(-\infty) = 0$, $F(\infty) = 1$
- $F(x)$ gives the total probability accumulated up to $x$
- For continuous distributions:  
  $$
  P(a < X \leq b) = F(b) - F(a)
  $$



### 🔍 Example: Standard Normal Distribution

If $X \sim N(0, 1)$, then:

$$
P(X \leq 1) = F(1) \approx 0.8413
$$

This means there's **84.13% probability** that a standard normally distributed variable is $\leq 1$.



### 📌 Applications

- Probabilities over intervals  
- Tail probabilities  
- Statistical thresholding  
- Hypothesis testing  


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the standard normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

# Set the point up to which we calculate cumulative probability
x_val = 1
x_fill = np.linspace(-4, x_val, 1000)
pdf_fill = norm.pdf(x_fill)

# Plot PDF with shaded AUC
plt.figure(figsize=(12, 5))

# --- PDF with shaded AUC
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.fill_between(x_fill, pdf_fill, alpha=0.3, color='lightblue', label=f'AUC up to x={x_val}')
plt.title('Area Under the PDF (Cumulative Probability)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

# --- CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.axvline(x_val, color='gray', linestyle='--')
plt.axhline(norm.cdf(x_val), color='gray', linestyle='--')
plt.scatter([x_val], [norm.cdf(x_val)], color='red')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('P(X ≤ x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

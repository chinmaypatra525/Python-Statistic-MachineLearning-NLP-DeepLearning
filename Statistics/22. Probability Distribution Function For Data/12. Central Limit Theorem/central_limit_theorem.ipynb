## ðŸ§  What is the Central Limit Theorem?

The **Central Limit Theorem (CLT)** states that:

> *When independent random variables are added, their normalized sum tends toward a normal (Gaussian) distribution, even if the original variables themselves are not normally distributed.*

This holds true **as the sample size becomes large (n â†’ âˆž)**.



## ðŸ“ˆ Formal Statement

If \( X_1, X_2, \ldots, X_n \) are i.i.d. (independent and identically distributed) random variables with:

- Mean: \( \mu \)
- Standard deviation: \( \sigma \)

Then the sampling distribution of the sample mean:

\[
\bar{X} = \frac{X_1 + X_2 + \cdots + X_n}{n}
\]

will approach a **normal distribution**:

\[
\bar{X} \sim N\left(\mu, \frac{\sigma^2}{n}\right)
\]

as \( n \to \infty \).



## âœ… Assumptions

- The variables are independent
- The sample size is sufficiently large (usually \( n \geq 30 \))
- The population has a finite mean and variance



## ðŸ“Š Real-Life Use Cases

| Use Case | Description |
|----------|-------------|
| A/B Testing | Used to estimate conversion rates from sample data |
| Manufacturing | Checking average defect rates across different batches |
| Finance | Portfolio returns from multiple assets over time |
| Biology | Average blood pressure readings from a sample group |



## ðŸ“Œ Key Insight

Even if the underlying data is **not normally distributed**, the **distribution of the sample mean** becomes **approximately normal** as the sample size increases.



## ðŸ“‰ Python Code

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate non-normal population (Exponential distribution)
population = np.random.exponential(scale=2, size=100000)

# Plot histogram of the population
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black')
plt.title("Original Population (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Simulate sampling distribution
sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

# Plot histogram of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, color='lightgreen', edgecolor='black')
plt.title("Sampling Distribution of Sample Means (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

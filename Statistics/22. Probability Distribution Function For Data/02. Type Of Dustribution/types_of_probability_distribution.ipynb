## 01. Bernoulli Distribution

### üß† What is Bernoulli Distribution?

A **Bernoulli Distribution** is the simplest type of probability distribution used for **binary outcomes** ‚Äî it models events with exactly two possible outcomes:

- ‚úÖ Success (1)  [with probability p]
- ‚ùå Failure (0)  [with probability 1-p]



### üßæ Real-Life Examples

| Scenario                        | Success (1) | Failure (0) |
|---------------------------------|-------------|-------------|
| Coin toss (Heads = success)     | Heads       | Tails       |
| Will it rain today?             | Yes         | No          |
| Email is spam?                  | Spam        | Not spam    |
| Customer makes a purchase?      | Yes         | No          |
| Machine working                 | Working     | Not Working |



### üìä Properties

| Property           | Description                                 |
|--------------------|---------------------------------------------|
| **Type**           | Discrete                                    |
| **Possible Values**| 0 and 1                                     |
| **Parameter**      | \( p \in [0, 1] \), probability of success   |
| **Mean (Œº)**       | \( p \)                                     |
| **Variance (œÉ¬≤)**  | \( p(1 - p) \)                               |
| **Support**        | \( x \in \{0, 1\} \)                         |



### üìä Summary

| Feature              | Description                        |
|----------------------|------------------------------------|
| **Distribution Type**| Discrete                           |
| **Possible Outcomes**| 0, 1                               |
| **Parameter**        | \( p \) (probability of success)   |
| **Mean**             | \( p \)                            |
| **Variance**         | \( p(1 - p) \)                      |
| **Example Use Case** | Coin toss, purchase decision, etc. |



### üí° Key Insights

- ‚úÖ **Only one trial**: yes/no, win/lose, etc.
- üìä Useful for **binary classification**, success/failure models
- ü§ñ Used in **machine learning**, decision trees, and logistic regression



### üî¢ Probability Mass Function (PMF)

The PMF for a Bernoulli random variable \( X \) is:

\[
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0, 1\}
\]

Alternatively:

- \( P(X = 1) = p \)  
- \( P(X = 0) = 1 - p \)



#### üîç Explanation of the Terms

| Symbol | Meaning |
|--------|---------|
| \( X \) | Random variable that can take value 0 (failure) or 1 (success) |
| \( x \) | Actual value of \( X \) (either 0 or 1) |
| \( p \) | Probability of success (i.e., \( P(X = 1) \)) |
| \( 1 - p \) | Probability of failure (i.e., \( P(X = 0) \)) |



#### How It Works

The PMF is designed so that:

- If \( x = 1 \):

\[
P(X = 1) = p^1 (1 - p)^0 = p
\]

- If \( x = 0 \):

\[
P(X = 0) = p^0 (1 - p)^1 = 1 - p
\]




#### Summary

So this **single formula**:

\[
P(X = x) = p^x (1 - p)^{1 - x}
\]

covers **both cases (success and failure)** compactly and elegantly!




### üõ† Python Code:

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

p = 0.3
x = [0, 1]
probs = bernoulli.pmf(x, p)

plt.bar(x, probs, color='purple')
plt.title('Bernoulli Distribution (p=0.3)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1])
plt.grid(True)
plt.show()
```

## 02. Binomial Distribution

### üß† What is Binomial Distribution?

A **Binomial Distribution** models the number of **successes** in a fixed number of **independent trials**, each with the same probability of success.

Think of it as **repeated Bernoulli trials**.

‚úÖ Each trial has only two outcomes: **success** or **failure**  
‚úÖ All trials are independent  
‚úÖ Probability of success (**p**) remains the same in each trial



### üßæ Real-Life Examples

| Scenario                                                      | Success   | Failure   |
|---------------------------------------------------------------|-----------|-----------|
| Flipping a coin 10 times                                      | Heads     | Tails     |
| Sending 100 emails and checking how many get replies          | Reply     | No reply  |
| Quality control: Checking 20 items for how many are defective | Defective | Good      |


### ‚úÖ Use Cases

| Domain              | Use Case                                                                 |
|---------------------|-------------------------------------------------------------------------|
| Quality Control      | Checking how many products in a batch are defective                    |
| Marketing            | Estimating how many customers will respond to a promotional email      |
| Medicine             | Predicting how many patients out of 100 will respond to a new drug     |
| Education            | Estimating how many students pass an exam with a fixed pass rate       |
| Survey Analysis      | Calculating probability of X people agreeing out of N sampled          |
| Cybersecurity        | Probability of system failure after repeated login attempts             |



### üìä Parameters and Properties

| Property              | Value / Formula              |
|-----------------------|------------------------------|
| **Distribution Type** | Discrete                     |
| **Random Variable**   | \( X \in \{0, 1, ..., n\} \) |
| **Number of Trials**  | \( n \)                      |
| **Success Probability**| \( p \)                     |
| **Failure Probability**| \( 1 - p \)                 |
| **Mean (Œº)**          | \( np \)                     |
| **Variance (œÉ¬≤)**     | \( np(1 - p) \)              |
| **Support**           | \( x = 0, 1, 2, ..., n \)    |



### üìä Summary

| Feature             | Description                                             |
|---------------------|---------------------------------------------------------|
| **Distribution Type** | Discrete                                              |
| **Parameters**        | \( n \): number of trials, \( p \): probability of success |
| **Mean**              | \( np \)                                              |
| **Variance**          | \( np(1 - p) \)                                       |
| **Shape**             | Symmetrical if \( p = 0.5 \), skewed otherwise        |
| **Common Use Cases**  | Surveys, Email Campaigns, Quality Testing, Exams      |



### üí° Key Insights

üìä **Binomial** is based on **Bernoulli**, but repeated over \( n \) trials.

üìà The **shape** of the distribution depends on both \( n \) and \( p \):

- Symmetric if \( p = 0.5 \)  
- Skewed right if \( p < 0.5 \)  
- Skewed left if \( p > 0.5 \)

üîç Used extensively in **quality control**, **A/B testing**, **clinical trials**, and **survey analysis**.

üß† If \( n \) is large and \( p \) is not too close to 0 or 1, it **approximates the Normal distribution**.




### üî¢ Probability Mass Function (PMF)

The PMF for the Binomial distribution is:

\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\]

Where:  
- \( n \) = number of trials  
- \( k \) = number of successes  
- \( p \) = probability of success  
- \( \binom{n}{k} = \frac{n!}{k!(n - k)!} \) is the combination formula



### üõ† Python Code:

```python
from scipy.stats import binom

n, p = 10, 0.5
x = range(n+1)
probs = binom.pmf(x, n, p)

plt.bar(x, probs, color='blue')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('k (successes)')
plt.ylabel('P(X = k)')
plt.grid(True)
plt.show()
```

## 03. Poisson Distribution

### üß† What is Poisson Distribution?

The **Poisson Distribution** models the probability of a given number of events occurring in a **fixed interval of time or space**, given that these events occur **independently** and at a **constant average rate**.

> üìå It is useful when you're counting occurrences of **rare events**.



### ‚úÖ Key Characteristics

- Events are **independent** of each other
- Events happen at a **constant average rate** \( \lambda \)
- Two events can't happen at **exactly the same time**
- The probability of an event occurring in a small interval is proportional to the **length of the interval**



### üßæ Real-Life Examples

| Scenario                             | Event Counted                           |
|--------------------------------------|------------------------------------------|
| Number of emails received per hour   | Emails                                   |
| Number of earthquakes in a region per year | Earthquakes                         |
| Number of server crashes in a day    | Server crashes                           |
| Number of customer arrivals per minute | Arrivals at a service counter         |



### üíº Real-Life Use Cases

| Domain           | Use Case                                                                 |
|------------------|--------------------------------------------------------------------------|
| Customer Support | Number of support tickets received per hour                              |
| Network Traffic  | Packets arriving at a router per second                                  |
| Retail           | Number of customers entering a store per minute                          |
| Website          | Number of clicks on an ad per hour                                       |
| Logistics        | Number of delivery requests per day in a region                          |
| Insurance        | Number of claims filed in a month                                        |
| Call Center      | Incoming calls received per 10-minute interval                           |
| Biology          | Mutations in a DNA strand over a certain length    



### üìê Parameters and Properties

| Property               | Value/Description               |
|------------------------|----------------------------------|
| Type                   | Discrete                         |
| Parameter              | \( \lambda > 0 \) (mean rate)   |
| Random Variable        | \( X \in \{0, 1, 2, ...\} \)     |
| Mean (Œº)               | \( \lambda \)                    |
| Variance (œÉ¬≤)          | \( \lambda \)                    |
| Support                | \( x = 0, 1, 2, \dots \)          |



### üìä Summary

| Feature              | Description                                      |
|----------------------|--------------------------------------------------|
| Distribution Type    | Discrete                                         |
| Random Variable      | \( X = 0, 1, 2, \dots \)                         |
| Parameter            | \( \lambda \): average rate of success (events/unit) |
| Mean (Œº)             | \( \lambda \)                                    |
| Variance (œÉ¬≤)        | \( \lambda \)                                    |
| Support              | \( x \in \mathbb{N}_0 \)                         |
| Used For             | Counting rare events in a fixed interval         |




### üí° Key Insights

‚úÖ The **Poisson distribution** is often used to model **rare events**.

‚è±Ô∏è It assumes events occur **independently** and at a **constant average rate**.

üî¢ **Mean and Variance are equal**:  
\[
\mu = \sigma^2 = \lambda
\]

üìâ The distribution is **right-skewed for small** \( \lambda \); becomes more **symmetric as \( \lambda \) increases**.

üîÅ **Poisson is a limiting case** of the Binomial Distribution as:  
\[
n \to \infty,\quad p \to 0,\quad \text{and} \quad np = \lambda
\]



### üî¢ Probability Mass Function (PMF)

The PMF of a Poisson-distributed random variable \( X \) is:

\[
P(X = x) = \frac{\lambda^x e^{-\lambda}}{x!}, \quad x = 0, 1, 2, \dots
\]

Where:
- \( \lambda \) = average number of events
- \( e \) = Euler‚Äôs number (~2.718)
- \( x! \) = factorial of \( x \)





### üõ† Python Code:


```python
from scipy.stats import poisson

lambda_ = 3
x = range(10)
probs = poisson.pmf(x, lambda_)

plt.bar(x, probs, color='green')
plt.title('Poisson Distribution (Œª=3)')
plt.xlabel('k (events)')
plt.ylabel('P(X = k)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lam = 3  # Average number of events

# X values
x = np.arange(0, 11)
pmf = poisson.pmf(x, mu=lam)

# Plot
plt.bar(x, pmf, color='skyblue')
plt.title(f"Poisson Distribution (Œª={lam})")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True, axis='y')
plt.show()

# Print PMF values
for val, prob in zip(x, pmf):
    print(f"P(X = {val}) = {prob:.4f}")
```

## 04. Normal (Gaussian) Distribution

### üß† What is Normal (Gaussian) Distribution?

The **Normal Distribution** (also called **Gaussian Distribution**) is a **continuous probability distribution** that is symmetric about the mean. It describes many natural phenomena such as height, weight, test scores, etc.

> üìå It is the most important distribution in statistics because of the **Central Limit Theorem (CLT)**.



### ‚úÖ Key Characteristics

- **Symmetrical bell-shaped curve**
- **Mean = Median = Mode**
- Determined by **two parameters**:
  - Mean \( \mu \): Center of the distribution
  - Standard deviation \( \sigma \): Spread/width of the curve
- Total area under the curve = 1



### üßæ Real-Life Examples

| Scenario                          | Description                            |
|-----------------------------------|----------------------------------------|
| Human heights                     | Most people are of average height      |
| IQ scores                         | Bell curve distribution around 100     |
| Measurement errors                | Small errors cluster near the mean     |
| Exam scores in large class        | Often normally distributed             |



### üíº Real-Life Use Cases

| Domain           | Use Case                                                 |
|------------------|-----------------------------------------------------------|
| Education        | Distribution of students' test scores                     |
| Manufacturing    | Measurement errors in production                          |
| Healthcare       | Blood pressure levels among a population                  |
| Finance          | Daily stock price returns                                 |
| Psychology       | IQ scores distribution                                    |
| Agriculture      | Weight of fruits or crops                                 |


### üìê Parameters and Properties

| Property         | Description                                |
|------------------|--------------------------------------------|
| Type             | Continuous                                 |
| Mean (Œº)         | Center of the distribution                 |
| Standard Deviation (œÉ) | Controls the spread (width)          |
| Support          | \( x \in (-\infty, +\infty) \)             |
| Skewness         | 0 (Perfectly symmetrical)                  |
| Kurtosis         | 3 (Mesokurtic)                             |



### üìä Summary

| Feature              | Description                               |
|----------------------|-------------------------------------------|
| Distribution Type    | Continuous                                |
| Shape                | Symmetrical bell curve                    |
| Parameters           | Œº (mean), œÉ (standard deviation)          |
| Support              | \( x \in (-\infty, \infty) \)             |
| Mean (Œº)             | Center of the curve                       |
| Variance (œÉ¬≤)        | Spread or width of the curve              |
| Skewness             | 0 (perfect symmetry)                      |
| Kurtosis             | 3 (mesokurtic - normal peak)              |



### üí° Key Insights

‚úÖ The **mean, median, and mode are all equal** in a normal distribution.

üìâ According to the **Empirical Rule**:
- ~68% of data falls within ¬±1œÉ
- ~95% within ¬±2œÉ
- ~99.7% within ¬±3œÉ

üîÅ **Many distributions (e.g., Binomial, Poisson)** approximate the Normal Distribution under large sample sizes via the **Central Limit Theorem**.

üåç **Most real-world phenomena** (e.g., height, test scores, measurement errors) follow or approximate a **Normal Distribution** due to the aggregation of many small random effects.



### üî¢ Probability Density Function (PDF)

The PDF of the normal distribution is:

\[
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
\]

Where:
- \( \mu \): Mean
- \( \sigma \): Standard deviation
- \( e \): Euler‚Äôs number (~2.718)



### üõ† Python Code:

```python
from scipy.stats import norm
import numpy as np

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, color='red')
plt.fill_between(x, y, alpha=0.3, color='salmon')
plt.title('Normal Distribution (Œº=0, œÉ=1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0     # Mean
sigma = 1  # Standard Deviation

# X values
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, pdf, label=f'N({mu}, {sigma}¬≤)', color='blue')
plt.fill_between(x, pdf, color='lightblue', alpha=0.5)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()
```

## 05. Standard Normal Distribution & Z Score

### üß† What is Standard Normal Distribution?

The **Standard Normal Distribution** is a special case of the **Normal Distribution** with:

- Mean (Œº) = 0  
- Standard Deviation (œÉ) = 1

It is used to simplify calculations and comparisons across different normal distributions by converting values to a **Z-score**.



### ‚úÖ Why Use Standard Normal?

Standardizing a normal distribution lets you:
- Compare scores from different distributions
- Use **Z-tables** or `scipy.stats.norm.cdf` to find probabilities



### üßÆ What is a Z-Score?

A **Z-score** tells us **how many standard deviations** a data point is from the mean.

\[
Z = \frac{X - \mu}{\sigma}
\]

- \( X \): is the original data point  
- \( \mu \): is the mean  
- \( \sigma \): is the standard deviation  
- \( Z \): standardized score (Z-score)



### üìä Interpretation of Z-Scores

| Z-Score | Meaning                       |
|---------|-------------------------------|
| 0       | Exactly at the mean           |
| +1      | 1 SD above the mean           |
| -1      | 1 SD below the mean           |
| +2      | 2 SDs above the mean          |
| -2      | 2 SDs below the mean          |
| > +3 or < -3 | Extreme outliers         |



### üî¢ Key Probabilities in Z-Scores

| Z-Score Range      | Area Under Curve (Approx.) |
|--------------------|----------------------------|
| -1 < Z < 1         | 68.27%                     |
| -2 < Z < 2         | 95.45%                     |
| -3 < Z < 3         | 99.73%                     |



### üßæ Real-Life Examples

| Scenario                                         | What Z-Score Represents                  |
|--------------------------------------------------|-------------------------------------------|
| Exam scores comparison between different tests   | Relative performance                      |
| Stock market returns                             | Measuring volatility                      |
| Quality control in manufacturing                 | Identifying defects beyond acceptable range |
| Detecting outliers in data                       | Extreme deviations from the mean          |



### üìä Summary

| Property            | Value                |
|---------------------|----------------------|
| Distribution Type   | Continuous           |
| Mean (Œº)            | 0                    |
| Standard Deviation (œÉ) | 1                 |
| Support             | \( x \in (-\infty, \infty) \) |
| PDF Formula         | \( f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2} \) |
| Symmetry            | Symmetric about 0    |
| Total Area          | 1                    |



### üí° Key Insights

üßÆ **Z-score** helps compare data across different **normal distributions**.

üìà **Standard Normal Distribution** is the reference model used in statistics.

üìä Widely used in:
- Hypothesis testing  
- Confidence intervals  
- Outlier detection

üéØ If ***Z-score > 2*** or ***Z-score < -2***, the data point is significantly different from the mean.



### üõ† Python Code:

```python
z = np.linspace(-4, 4, 1000)
y = norm.pdf(z, 0, 1)

plt.plot(z, y, color='orange')
plt.title('Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# X range
x = np.linspace(-4, 4, 1000)

# Standard normal PDF
pdf = norm.pdf(x, loc=0, scale=1)

# Plot
plt.plot(x, pdf, color='blue', label='Standard Normal N(0,1)')
plt.fill_between(x, pdf, alpha=0.2, color='skyblue')
plt.title("Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Density")
plt.grid(True)
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.show()
```

### üìå Example: Convert to Z-Score

#### Example 1:

A student scores **85** in a test where:

- Mean \( \mu = 70 \)  
- Standard Deviation \( \sigma = 10 \)

#### Z-Score Formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

\[
Z = \frac{85 - 70}{10} = 1.5
\]

‚úÖ **Interpretation**: The score is **1.5 standard deviations above the mean**.


```python
from scipy.stats import norm

# P(Z < 1.5)
prob = norm.cdf(1.5)
print(f"P(Z < 1.5) = {prob:.4f}")  # Output: ~0.9332

# P(Z > 1.5)
print(f"P(Z > 1.5) = {1 - prob:.4f}")  # Output: ~0.0668
```

## 06. Uniform Distribution

### üß† What is Uniform Distribution?

The **Uniform Distribution** is a type of probability distribution in which **all outcomes are equally likely** within a defined range.

There are two types:
- **Discrete Uniform Distribution** (e.g., dice rolls)
- **Continuous Uniform Distribution** (e.g., random number between 0 and 1)



### üßæ Real-Life Examples

| Scenario                                 | Type         |
|------------------------------------------|--------------|
| Rolling a fair die (1‚Äì6)                 | Discrete     |
| Random number generated between 0 and 1  | Continuous   |
| Spinning a fair wheel                    | Continuous   |



### ‚úÖ Use Cases

- üé≤ **Random Sampling**  
  Generate binary outcomes like success/failure, heads/tails, etc.

- üß™ **Simulations**  
  Simulate real-world binary events, e.g., coin flips, power on/off, etc.

- üéÆ **Modeling Games and Lottery**  
  Use Bernoulli trials to simulate win/loss scenarios.

- ü§ñ **Initializing Weights in Machine Learning**  
  Randomly initialize binary weights or dropout masks in neural networks.


### üìä Properties of Continuous Uniform Distribution

| Property         | Description                        |
|------------------|------------------------------------|
| Type             | Continuous                         |
| Support          | \( x \in [a, b] \)                 |
| Mean (Œº)         | \( \mu = \frac{a + b}{2} \)        |
| Variance (œÉ¬≤)    | \( \sigma^2 = \frac{(b - a)^2}{12} \) |
| PDF              | \( f(x) = \frac{1}{b - a} \) for \( x \in [a, b] \) |
| CDF              | \( F(x) = \frac{x - a}{b - a} \) for \( x \in [a, b] \) |



### üìä Summary

| Property                  | Description                             |
|---------------------------|-----------------------------------------|
| Type                      | Discrete / Continuous                   |
| PDF (Continuous)          | \( f(x) = \frac{1}{b - a} \) for \( x \in [a, b] \) |
| PMF (Discrete)            | \( P(X = x) = \frac{1}{n} \) for \( n \) outcomes |
| Mean                      | \( \mu = \frac{a + b}{2} \)             |
| Variance                  | \( \sigma^2 = \frac{(b - a)^2}{12} \)   |
| Support                   | \( x \in [a, b] \)                      |
| Shape                     | Rectangular                             |



### üí° Key Insights

- üìä **All outcomes are equally likely.**
- üé≤ Common in **simulations**, **games of chance**, and **random sampling**.
- üìê The shape is a **rectangle** ‚Äî hence also called a **"rectangular distribution."**
- ‚öñÔ∏è Represents **maximum uncertainty**: no outcome is more likely than another.



### üî¢ Probability Density Function (PDF)

\[
f(x) = 
\begin{cases}
\frac{1}{b - a}, & a \le x \le b \\
0, & \text{otherwise}
\end{cases}
\]



### üîÑ Cumulative Distribution Function (CDF)

\[
F(x) = 
\begin{cases}
0, & x < a \\
\frac{x - a}{b - a}, & a \le x \le b \\
1, & x > b
\end{cases}
\]



### üõ† Python Code:

```python
from scipy.stats import uniform

a, b = 0, 1
x = np.linspace(a, b, 1000)
y = uniform.pdf(x, a, b - a)

plt.plot(x, y, color='brown')
plt.title('Uniform Distribution (0 to 1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Range [a, b]
a = 2
b = 10
x = np.linspace(0, 12, 1000)
pdf = uniform.pdf(x, loc=a, scale=b-a)

# Plot PDF
plt.plot(x, pdf, label="Uniform PDF", color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='skyblue')
plt.title("Continuous Uniform Distribution PDF (a=2, b=10)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.show()
```

## 07. Log Normal Distribution

### üß† What is Log-Normal Distribution?

A **Log-Normal Distribution** is a **continuous probability distribution** of a random variable whose logarithm is **normally distributed**.

In other words, if:

\[
Y = \ln(X) \sim \mathcal{N}(\mu, \sigma^2)
\]

Then:

\[
X \sim \text{Log-Normal}(\mu, \sigma^2)
\]



### üßæ Real-Life Examples

| Scenario                              | Description                        |
|---------------------------------------|------------------------------------|
| Stock prices                          | Prices can't be negative           |
| Income distribution                   | Right-skewed with long tail        |
| Biological measures                   | e.g., cell size, survival time     |
| Time to complete a task               | Often multiplicative factors       |



### ‚úÖ When to Use

- üìà **When the data is positively skewed**  
  Ideal for distributions like income, stock prices, etc.

- üö´ **When values can‚Äôt be negative**  
  Log-Normal distributions are defined only for positive real numbers.

- ‚úñÔ∏è‚ûï **When you're modeling multiplicative effects**  
  Useful when outcomes result from multiplying many small positive factors.



### üìä Properties

| Property           | Value / Formula                                         |
|--------------------|---------------------------------------------------------|
| Distribution Type  | Continuous                                              |
| Support            | \( x \in (0, \infty) \)                                 |
| PDF                | \( f(x) = \frac{1}{x\sigma\sqrt{2\pi}} e^{ -\frac{(\ln x - \mu)^2}{2\sigma^2} } \) |
| CDF                | \( F(x) = \Phi\left(\frac{\ln x - \mu}{\sigma}\right) \) |
| Mean (Œº)           | \( e^{\mu + \frac{\sigma^2}{2}} \)                      |
| Median             | \( e^\mu \)                                             |
| Mode               | \( e^{\mu - \sigma^2} \)                                |
| Variance (œÉ¬≤)      | \( (e^{\sigma^2} - 1)e^{2\mu + \sigma^2} \)             |
| Skewness           | Always > 0 (positively skewed)                         |



### üìä Summary

| Property            | Value                                 |
|---------------------|----------------------------------------|
| Distribution Type   | Continuous                             |
| Support             | \( x \in (0, \infty) \)                |
| Mean                | \( e^{\mu + \frac{\sigma^2}{2}} \)     |
| Median              | \( e^\mu \)                            |
| Mode                | \( e^{\mu - \sigma^2} \)               |
| Variance            | \( (e^{\sigma^2} - 1) e^{2\mu + \sigma^2} \) |
| Skewness            | Positive (right-skewed)                |
| Parameters          | \( \mu \): mean of log(X), \( \sigma \): std dev of log(X) |



### üí° Key Insights

- üìà **Log-Normal** is useful when values **cannot be negative** and **grow multiplicatively**.
- ‚ÜóÔ∏è It is **not symmetric** like the Normal Distribution ‚Äî it has a **long tail on the right**.
- üîÑ Taking the **log** of a log-normal variable gives you a **normal distribution**.
- üíº Commonly used in:
  - **Finance** (e.g., stock prices)
  - **Biology** (e.g., reaction times)
  - **Reliability engineering** (e.g., time to failure)



### üî¢ Probability Density Function (PDF)

\[
f(x; \mu, \sigma) = 
\begin{cases}
\frac{1}{x \sigma \sqrt{2\pi}} \exp\left( - \frac{(\ln x - \mu)^2}{2\sigma^2} \right), & x > 0 \\
0, & x \leq 0
\end{cases}
\]

## 08. Power Law Distribution

### üß† What is Power Law Distribution?

The **Power Law Distribution** is a **heavy-tailed probability distribution** that follows the form:

\[
P(x) \propto x^{-\alpha}
\]

It describes events where **large occurrences are rare**, but **small occurrences are common**, and the distribution **does not decay exponentially** like normal distributions.



### üßæ Real-Life Examples

| Scenario                               | Description                             |
|----------------------------------------|-----------------------------------------|
| Earthquake magnitudes                  | Many small quakes, few large ones       |
| Word frequency in natural languages    | Some words used very often, others rarely |
| City populations                       | Few large cities, many small ones       |
| Wealth distribution                    | Small number of people own majority     |
| Internet link structure                | Few hubs, many low-degree nodes         |



### ‚úÖ When to Use

- üìä **When data is right-skewed and follows scale-invariance**  
  Power law distributions apply where patterns repeat at every scale (e.g., city sizes, file sizes).

- ü¶¢ **When rare but large events dominate (e.g., black swans)**  
  Useful for modeling outliers that significantly impact the system.

- üåê **Useful in complex systems, networks, finance, and geophysics**  
  Examples include earthquake magnitudes, wealth distribution, internet traffic, and stock returns.



### üìä Properties

| Property           | Value / Formula                                           |
|--------------------|-----------------------------------------------------------|
| Distribution Type  | Discrete or Continuous                                    |
| Support            | \( x \geq x_{min} > 0 \)                                  |
| PDF (Continuous)   | \( f(x) = \frac{\alpha - 1}{x_{min}} \left( \frac{x}{x_{min}} \right)^{-\alpha} \) |
| PMF (Discrete)     | \( P(X=x) = \frac{x^{-\alpha}}{\zeta(\alpha)} \) (zeta = Riemann Zeta function) |
| Mean (Œº)           | Defined for \( \alpha > 2 \), \( \mu = \frac{\alpha - 1}{\alpha - 2} x_{min} \) |
| Variance (œÉ¬≤)      | Defined for \( \alpha > 3 \) |
| Skewness           | Heavy-tail, high skew                                     |



### üìä Summary

| Property            | Value                                       |
|---------------------|----------------------------------------------|
| Type                | Continuous or Discrete                      |
| Support             | \( x \geq x_{\text{min}} \)                 |
| PDF                 | \( f(x) = C x^{-\alpha} \), \( \alpha > 1 \) |
| Parameters          | \( \alpha \): shape (scaling) parameter     |
| Mean (if \( \alpha > 2 \))     | \( \frac{\alpha - 1}{\alpha - 2} x_{\text{min}} \) |
| Variance (if \( \alpha > 3 \)) | \( \frac{(\alpha - 1)x_{\text{min}}^2}{(\alpha - 3)(\alpha - 2)^2} \) |
| Skewness            | Right-skewed (heavy-tailed)                 |



### üí° Key Insights

- üìè **Power law** models "scale-free" phenomena, where **large values are rare but possible**.
- üåÄ The **long tail** represents **high-impact, low-probability events**.
- üåç Used in modeling:
  - **Natural disasters**
  - **Network science**
  - **Econophysics**
- ‚ùó Often confused with **exponential** or **normal** distributions, but **power laws decay more slowly**.



### üîç Notes

- üìâ Power laws are typically visualized using **log-log plots**, where the curve becomes **linear**.
- üìä **Verifying** a true power-law relationship requires **rigorous statistical testing** (e.g., Clauset-Shalizi-Newman method).



### üî¢ Probability Density Function (PDF)

\[
f(x; \alpha, x_{min}) =
\begin{cases}
\frac{\alpha - 1}{x_{min}} \left( \frac{x}{x_{min}} \right)^{-\alpha}, & x \geq x_{min} \\
0, & x < x_{min}
\end{cases}
\]

- \( \alpha > 1 \): shape parameter
- \( x_{min} \): minimum value


### üõ† Python Code:

```python
x = np.linspace(1, 100, 1000)
alpha = 2.5
y = x ** (-alpha)

plt.plot(x, y, color='black')
plt.title('Power Law Distribution (Œ±=2.5)')
plt.xlabel('x')
plt.ylabel('P(X)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 2.5
xmin = 1
x = np.linspace(xmin, 50, 1000)

# PDF formula
pdf = (alpha - 1) / xmin * (x / xmin) ** (-alpha)

# Plot
plt.plot(x, pdf, label=f"Œ±={alpha}, xmin={xmin}", color='blue')
plt.title("Power Law Distribution (PDF)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.show()
```

## 09. Pareto Distribution


### üß† What is Pareto Distribution?

The **Pareto Distribution** is a **power-law probability distribution** used to describe many types of **observable phenomena in economics, finance, and social sciences**. It is named after **Vilfredo Pareto**, who observed that 80% of wealth was owned by 20% of the population ‚Äî known as the **80/20 rule**.

It models the probability that a variable (like wealth, file size, city population) is **greater than a certain value**.



### üßæ Real-Life Examples

| Scenario                                | Description                          |
|-----------------------------------------|--------------------------------------|
| Wealth distribution                     | Small % of people own most wealth    |
| File sizes on the internet              | Few large files, many small ones     |
| City populations                        | Few megacities, many small towns     |
| Insurance claim sizes                   | Few large claims, many small ones    |
| Word frequencies in language            | Few words used frequently            |



### ‚úÖ When to Use

- üìè When modeling systems with **scale invariance**
- üê≥ When data exhibits **long tail** or **rich-get-richer behavior**
- üíº Applicable in **economics, ecology, geophysics, insurance**, etc.



### üìä Properties

| Property               | Formula / Description                                               |
|------------------------|---------------------------------------------------------------------|
| Type                   | Continuous                                                          |
| Support                | \( x \geq x_m \) (minimum value)                                   |
| Shape Parameter        | \( \alpha > 0 \)                                                    |
| PDF                    | \( f(x) = \frac{\alpha x_m^\alpha}{x^{\alpha + 1}} \)              |
| CDF                    | \( F(x) = 1 - \left( \frac{x_m}{x} \right)^\alpha \)               |
| Mean                   | \( \mu = \frac{\alpha x_m}{\alpha - 1} \), if \( \alpha > 1 \)      |
| Variance               | \( \sigma^2 = \frac{\alpha x_m^2}{(\alpha - 1)^2 (\alpha - 2)} \), if \( \alpha > 2 \) |
| Skewness               | Heavy right-skewed                                                 |
| Memoryless Property    | No                                                                 |



### üß† Summary

| Feature        | Value |
|----------------|-------|
| **Distribution** | Pareto |
| **PDF**          | \( f(x) = \frac{\alpha x_m^\alpha}{x^{\alpha+1}} \) |
| **CDF**          | \( F(x) = 1 - \left(\frac{x_m}{x}\right)^\alpha \) |
| **Support**      | \( x \geq x_m \) |
| **Skewness**     | Heavy right-tail |
| **Applications** | Wealth, size distribution, file sizes |



### üí° Key Insights

- üí∞ Models **"rich-get-richer"** and **heavy-tail** phenomena.
- üåê Often used in:
  - **Economics** (e.g., income distribution)
  - **Network modeling**
  - **Risk assessment**
- ‚ö†Ô∏è Unlike Normal Distribution, **Pareto has infinite variance if Œ± ‚â§ 2**.
- üîó Closely related to the **Power Law Distribution**, but with a **specific lower bound**.



### üîç Notes

- üìâ When plotted on a **log-log scale**, Pareto forms a **straight line** (indicative of power-law behavior).
- ‚ùó Useful in scenarios where **mean and variance may not exist** (for low values of Œ±).



### üî¢ Probability Density Function (PDF)

\[
f(x; \alpha, x_m) =
\begin{cases}
\frac{\alpha x_m^\alpha}{x^{\alpha + 1}}, & x \geq x_m \\
0, & x < x_m
\end{cases}
\]

Where:
- \( \alpha \) = shape parameter
- \( x_m \) = scale parameter (minimum value)



### üìå Interpretation

- üîΩ **Larger Œ± ‚Üí faster drop-off** (less extreme events)  
- üìà **Smaller Œ± ‚Üí heavier tail** (more extreme values)  
- ‚ôæÔ∏è **Pareto explains self-similarity**, where distributions look the same on different scales.



### üõ† Python Code:

```python
from scipy.stats import pareto

b = 2.62
x = np.linspace(1, 5, 1000)
y = pareto.pdf(x, b)

plt.plot(x, y, color='magenta')
plt.title('Pareto Distribution (Œ±=2.62)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
